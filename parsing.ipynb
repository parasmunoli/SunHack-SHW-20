```python
import PyPDF2

def extract_text_from_pdf(pdf_file_path, output_text_file):
    with open(pdf_file_path, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        
        # Initialize an empty string to store the extracted text
        extracted_text = ''
        
        # Loop through each page in the PDF
        for page_num in range(len(pdf_reader.pages)):
            # Extract text from the current page
            page = pdf_reader.pages[page_num]
            extracted_text += page.extract_text()

    # Write the extracted text to a text file
    with open(output_text_file, 'w', encoding='utf-8') as text_file:
        text_file.write(extracted_text)
    
    return extracted_text

# Path to the PDF file you want to extract text from
pdf_file_path = 'resume/temp.pdf'

# Path to the text file where you want to save the extracted text
output_text_file = 'output.txt'

# Call the function to extract text from the PDF and save it to a text file
res = extract_text_from_pdf(pdf_file_path, output_text_file)
print(res)
```

```python
import PyPDF2

def extract_text_from_pdf(pdf_file_path, output_text_file):
    with open(pdf_file_path, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        
        # Initialize an empty list to store the extracted text from each page
        extracted_text_list = []
        
        # Loop through each page in the PDF
        for page_num in range(len(pdf_reader.pages)):
            # Extract text from the current page
            page = pdf_reader.pages[page_num]
            extracted_text_list.append(page.extract_text())

    # Join the extracted text from all pages with spaces
    extracted_text = ' '.join(extracted_text_list)
    
    # Write the extracted text to a text file
    with open(output_text_file, 'w', encoding='utf-8') as text_file:
        text_file.write(extracted_text)
    
    return extracted_text

# Path to the PDF file you want to extract text from
pdf_file_path = 'resume/temp.pdf'

# Path to the text file where you want to save the extracted text
output_text_file = 'output.txt'

# Call the function to extract text from the PDF and save it to a text file
res = extract_text_from_pdf(pdf_file_path, output_text_file)
print(res)
```
